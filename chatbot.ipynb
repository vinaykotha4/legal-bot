```python
!pip install langchain
!pip install gradio
!pip install huggingface_hub

import os
import gradio as gr
from langchain.chat_models import ChatOpenAI
from langchain import LLMChain, PromptTemplate
from langchain.memory import ConversationBufferMemory

# Set OpenAI API key directly
OPENAI_API_KEY = "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"

# Template for conversation prompt
template = """
You are a legal assistant providing legal advice.
{chat_history}
User: {user_message}
Chatbot:
"""
prompt = PromptTemplate(input_variables=["chat_history", "user_message"], template=template)
# Conversation memory
memory = ConversationBufferMemory(memory_key="chat_history")

# Initialize language model chain
llm_chain = LLMChain(
    llm=ChatOpenAI(temperature='0.5', model_name="gpt-3.5-turbo", openai_api_key=OPENAI_API_KEY),
    prompt=prompt,
    verbose=True,
    memory=memory
)

def is_legal_related(query):
    # Legal-related keywords
    legal_keywords = [
        "law", "legal", "legal advice", "attorney", "lawyer",
        "court", "case", "lawsuit", "legal document", "contract",
        "legal rights", "legal obligations", "legal procedure", "legal system",
        "civil law", "criminal law", "family law", "business law", "property law",
        "intellectual property law", "employment law", "constitutional law", "tort law",
        "legal dispute", "legal precedent", "legal consultation", "legal representation",
        "legal jurisdiction", "legal code", "legal framework", "legal terminology",
        "legal entity", "legal responsibility", "legal liability", "legal practice",
        "legal ethics", "legal profession", "legal proceedings", "legal interpretation",
        "legal contract", "legal agreement", "legal remedy", "legal settlement",
        "legal issue", "legal matter", "legal assistance", "legal counsel",
        "legal question", "legal aid", "legal research", "legal review",
        "legal argument", "legal opinion", "legal stance", "legal strategy",
        "legal procedure", "legal assistance", "legal representation", "legal defense",
        "legal offense", "legal compliance", "legal requirement", "legal violation",
        "legal compliance", "legal obligation", "legal responsibility"
    ]
    for keyword in legal_keywords:
        if keyword in query:
            return True
    return False

def get_legal_response(user_message, chat_history):
    if is_legal_related(user_message):
        # Provide legal-related response
        response = llm_chain.predict(user_message=user_message)
    else:
        response = "Sorry, I can only provide legal advice."
    return response

# Create chat interface
demo = gr.Interface(get_legal_response,
                    inputs=["text", "text"],
                    outputs="text",
                    title="Legal support chatbot",
                    description="Ask legal questions using keywords provided",
                    examples=[["What should I do if I received a legal notice?", ""],
                              ["What are my rights if my landlord wants to evict me?", ""],
                              ["Can you explain the process of trademark registration?", ""]])

# Launching the interface locally
demo.launch()
```
